# Prediction of weight of body, given brain weight

## Import Libraries

```
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import math
```

## Read Data Using Pandas

```
dataframe = pd.read_fwf('brain_body.txt')
x_values = dataframe[['Brain']].as_matrix()
y_values = dataframe[['Body']].as_matrix()
```

## Split the data into train and test

```
leng = int(math.ceil(0.8*len(x_values)))

train_X = x_values[:leng]
train_Y = y_values[: leng]
test_X = x_values[leng:]
test_Y = y_values[leng:]
```

## Fitting the Data On A Linear Regression Model

```
body_reg = linear_model.LinearRegression()
body_reg.fit(train_X, train_Y)
```

## Plotting the regression line

```
print(body_reg.score(test_Y , body_reg.predict(test_X) ))
```
